```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
img=cv2.imread("contour.jpg",0)
plt.figure()
plt.imshow(img,cmap="gray")
plt.axis("off")
```

Contour yapma

(img,internal veya external contourları extrack edelir,yatay dikey ve çapraz bölümleri sıkıştırılması)

```python
contours,hierarch=cv2.findContours(img,cv2.RETR_CCOMP,cv2.CHAIN_APPROX_SIMPLE)
```

```python
external_countours=np.zeros(img.shape)
print(external_countours)
```

```python
internal_countours=np.zeros(img.shape)
print(internal_countours)
```

```python
for i in range(len(contours)):
    #external
    if hierarch[0][i][3]==-1:
        cv2.drawContours(external_countours,contours,i,255,-1)
    else: #internal
        cv2.drawContours(internal_countours,contours,i,255,-1)
```

```python
plt.figure()
plt.imshow(external_countours,cmap="gray")
plt.axis("off")
```

```python
plt.figure()
plt.imshow(internal_countours,cmap="gray")
plt.axis("off")
```
